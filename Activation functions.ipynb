```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## 1. A max bounded rectifier unit

```python
plt.axis([-2.0, 2.0, -2.0, 2.0])

a = [-2, -1, 0, 1, 2]
b = [0, 0, 0, 1, 1]
plt.plot(a, b)
plt.grid()
```

## 2. A non monotonous activation function

```python
plt.axis([-2.0, 2.0, -2.0, 2.0])

a = [-2, -1, 0, 1, 2]
b = [0, 0, 1, 0, 0]
plt.plot(a, b)
plt.grid()
```

## 3. A rectifier that ranges from -1 to 1

```python
plt.axis([-2.0, 2.0, -2.0, 2.0])

a = [-2, -1, 0, 1, 2]
b = [-1, -1, 0, 1, 1]
plt.plot(a, b)
plt.grid()
```

## 4. A linear function with a plateau around 0

```python
plt.axis([-2.0, 2.0, -2.0, 2.0])

a = [-2, -1, 0, 1, 2]
b = [-1, 0, 0, 0, 1]
plt.plot(a, b)
plt.grid()
```
